## Hypothesis Testing
R.A. Fisher proposed hypothesis testing based on the statistical significance. This method asks if the data is well outside the region where we would expect to see it under the null hypothesis, $H_0$. If so, then we reject the null hypothesis in favor of the alternative hypothesis, $H_A$.

The p-value is the probability that observed experiment results would occur if the null hypothesis is the truth.

To apply a test of hypotheses to our problem, here are the steps:
1. State the null hypothesis and alternative hypothesis
2. Estabilish p-value threshold; also known as the significance level: $\alpha$.
3. Conduct the experiement and record the results.
4. Calculate p-value.
5. Compare p-value to $\alpha$:
 * If p-value is greater, do not reject $H_0$.
 * If p-value is less, reject $H_0$ in favor of $H_A$.

### Example: 2010 and 2013 Traffic Accidents in Barcelona

**1. State the null hypothesis and alternative hypothesis**
    * $H_0$: The mean of traffic accidents is the same in 2010 and 2013
    * $H_A$: The mean of traffic accidents is different in 2010 and 2013

**2. Estabilish p-value threshold; also known as the significance level: $\alpha = 0.05$.**

**3. Conduct the experiement and record the results:**

```python
data = pd.read_csv("data/ACCIDENTS_GU_BCN_2010.csv", encoding='latin-1')
#Create a new column which is the date
data['Date'] = data['Dia de mes'].apply(lambda x : str(x)) + '-' +  \
               data['Mes de any'].apply(lambda x : str(x))
data2 = data['Date']
counts2010 =data['Date'].value_counts()
print('2010: Mean', counts2010.mean())

data = pd.read_csv("data/ACCIDENTS_GU_BCN_2013.csv", encoding='latin-1')
#Create a new column which is the date
data['Date'] = data['Dia de mes'].apply(lambda x : str(x)) + '-' +  \
               data['Mes de any'].apply(lambda x : str(x))
data2 = data['Date']
counts2013 = data['Date'].value_counts()
print('2013: Mean', counts2013.mean())
m = len(counts2010)
n = len(counts2013)
p = (counts2013.mean() - counts2010.mean())
print('mean difference: ', p)
print('standard deviation of 2013 sample:', counts2013.std())
```

**4. Calculate p-value**

To approximate the p-value, we can follow the following procedure:
1. Pool the distributions, generate samples with size $n$ and compute the difference in the mean.
2. Generate samples with size $n$ and compute the difference in the mean.
3. Count how many differences are larger than the observed one

```python
x = counts2010
y = counts2013
pool = np.concatenate([x,y])
np.random.shuffle(pool)

fig, ax = plt.subplots(1, 1, figsize=(12, 3))
plt.hist(pool, bins = 25, density = True)
plt.ylabel('Frequency')
plt.xlabel('Number of accidents')
plt.title("Pooled distribution")
plt.show()
```

```python
# Sample generation
N = 10000 # number of samples
diff = []
for i in range(N):
    p1 = [random.choice(pool) for _ in range(n)]
    p2 = [random.choice(pool) for _ in range(n)]
    diff.append((np.mean(p1)-np.mean(p2)))

fig, ax = plt.subplots(1, 1, figsize=(12, 3))
plt.hist(diff, bins = 50, density = True)
plt.ylabel('Frequency')
plt.xlabel('Difference in the mean')
plt.show()
```

```python
# counting how many differences are larger than the observed one
diff2 = np.array(diff)
w1 = np.where(diff2 > p)[0]      
print('{} samples are larger than the observed mean difference'.format(len(w1)))

pVal = len(w1)/float(N)
print('p-value(Simulation) = {} ({:.2%}), Difference = {}'.format(pVal, pVal, p))
if pVal<0.05:
    print('Reject the null hypothesis')
else:
    print('Fails to reject the null hypothesis')
```

**4. Calculate p-value in traditional way**

Given the standard deviation of the population, $\sigma$, the standard deviation of the sampling distribution (standard error) is:
$$ \sigma_{\bar x} = \frac{\sigma}{\sqrt{n}} $$

However, in practice $\sigma$ is not known, so we use standard deviation of the sample, $s$ to approximate it. 


For our example, we calculate following statistics:

$$
\begin{align}
\frac{\bar x - \mu_{\bar x}}{\sigma_{\bar x}} &= \frac{\bar x - \mu_{\bar x}}{\frac{\sigma}{\sqrt{n}}} \\
&= \frac{\bar x - \mu_{\bar x}}{\frac{s}{\sqrt{n}}} \\
&= \frac {25.909589 - 24.8109589}{\frac{9.108325}{\sqrt{365}}} \\
&= 2.304 \\
\end{align}
$$

To get p-value, use the following test statistics:
 - z-test if $\sigma$ is known or have large sample size $(n > 30)$.
 - Otherwise, use t-test statistic with degree of freedom $(n-1)$.
 
For more details about the theory behind the test statistics, checkout central limit theorem and sampling distribution demostration at:
http://onlinestatbook.com/stat_sim/sampling_dist/index.html

In our example, we use z-statistics because $n=365$ and we can obtain the p-value of $0.0107$ from the Z-table.

Therefore, we **reject** the null hypothesis because it is less than $\alpha = 0.05$.


### Precision vs Recall
The following three terminologies are all closely related.
* Precision and Recall
* Type I Error and Type II Error
* False Postives and False Negatives

When making decision to reject the null hypothesis, there are two possible errors. A type I error (false positives) is the incorrect rejection of a true null hypothesis. Typically in hypothesis testing, we set the probability threshold of Type I error to 5% when rejecting the null hypothesis. A type II error (false negatives) is the failure to reject a false null hypothesis.

$$Precision = \frac{True_{Postives}}{True_{Positives} + False_{Positives}}$$

$$Recall = \frac{True_{Postives}}{True_{Positives} + False_{Negatives}}$$

# Bayesian Posterior Inference
Bayesian inference is a method of statistical inference in which Bayes' theorem is used to update the probability for a hypothesis as more evidence or information becomes available. Bayesian inference is an important technique in statistics, and especially in mathematical statistics. Bayesian updating is particularly important in the dynamic analysis of a sequence of data. Bayesian inference has found application in a wide range of activities, including science, engineering, philosophy, medicine, sport, and law. In the philosophy of decision theory, Bayesian inference is closely related to subjective probability, often called "Bayesian probability".

Probabilities are used to measure the uncertainty we have about parameters, and Bayes' theorem is the mechanism to correctly update those probabilities in the light of new data, hopefully reducing our uncertainty.

## Bayes’ Rule
$$P(A\mid B)={\frac {P(B\mid A)\,P(A)}{P(B)}}$$
where $A$ and $B$ are events and $P(B)\neq0$.
* $P(A|B)$ is a conditional probability; the likelihood of event A occuring given that $B$ is true.
* $P(B|A)$ is also a conditional probability; the likehood of event B occuring given that $A$ is true.
* $P(A)$ and $P(B)$ are the probabilities of observing $A$ and $B$ independently of each other; known as the marginal probability.


## The Coin-Flipping Problem
We toss a coin a number of times and record how many heads and tails we get. Based on this data, the problem is to determine if the coin is fair? Or, more generally, how biased is the coin? This problem and solution can be extended to many real problems consist of binary mutually exclusive outcomes such as 0 or 1, positive or negative, odds or evens, spam or ham, safe or unsafe, healthy or unhealthy, and so on.

In order to estimate the bias of a coin, and in general to answer any questions in a Bayesian setting, we will need data and a probabilistic model.

Let $\theta$ be the bias of the coin. If $\theta=1$, then the coin will always land heads and if $\theta=0$, then the coin will land tails.

Let $y$ be the toal number of heads flipped for $N$ number of tosses in the experiement.

Then, according to Bayes' rule:
$$P(\theta\mid y)={\frac {P(y\mid\theta)\,P(\theta)}{P(y)}}$$

Assuming each coin toss is independent (i.e. one toss does not affect other tosses), the likelihood is the binomial distribution:
$$P(\theta\mid y)={\frac {N!}{N!(N-y)!}}\theta^y(1-\theta)^{N-y}$$
This is a discrete distribution returning the probability of getting $y$ heads out of $N$ tosses.

```python
n_params = [1, 2, 4]
p_params = [0.25, 0.5, 0.75]
x = np.arange(0, max(n_params)+1)
f, ax = plt.subplots(len(n_params), len(p_params), sharex=True, sharey=True, figsize=(12, 5))
for i in range(3):
    for j in range(3):
        n = n_params[i]
        p = p_params[j]
        y = stats.binom(n=n, p=p).pmf(x)
        ax[i,j].vlines(x, 0, y, colors='b', lw=5)
        ax[i,j].set_ylim(0, 1)
        ax[i,j].plot(0, 0, label="$n = {:3.0f}$\n$\\theta = {:3.2f}$".format(n, p), alpha=0)
        ax[i,j].legend(fontsize=12)
ax[2,1].set_xlabel('$y$', fontsize=14)
ax[1,0].set_ylabel('$p(y|\\theta)$', fontsize=14)
ax[0,0].set_xticks(x)
```

So if we know $\theta$, the binnomial distribution will tell us the expected outcome distribution. However, we do not know $\theta$ That is where Bayesian Inference comes in!

Choose *beta distribution* as the prior.
$$P(\theta)={\frac {\Gamma(\alpha + \beta)}{\Gamma(\alpha)\Gamma(\beta)}}\theta^{\alpha-1}(1-\theta)^{\beta-1}$$
* $\Gamma$ is the *gamma function* and ${\frac {\Gamma(\alpha + \beta)}{\Gamma(\alpha)\Gamma(\beta)}}$ is a normalization constant that ensures the distribution integrates to $1$.
* $\alpha > 0$
* $\beta > 0$
* $0\leq\theta\leq1$

### Why did we choose *beta distribution* as the prior? 
Because it has the some desired properties like $\theta$ is restricted between $0$ and $1$. Also, by adjusting $\alpha$ and $\beta$, beta distribution can mimic uniform distribution, nomal distribution and biased asymetrical distribution. See the shape of the beta distributions below for different values of $\alpha$ and $\beta$.

```python
params = [0.5, 1, 2, 3]
x = np.linspace(0, 1, 100)
f, ax = plt.subplots(len(params), len(params), sharex=True, sharey=True, figsize=(12, 6))
for i in range(4):
    for j in range(4):
        a = params[i]
        b = params[j]
        y = stats.beta(a, b).pdf(x)
        ax[i,j].plot(x, y)
        ax[i,j].plot(0, 0, label="$\\alpha$ = {:3.2f}\n$\\beta$ = {:3.2f}".format(a, b), alpha=0)
        ax[i,j].legend(fontsize=12)
        ax[3,j].set_xlabel('$\\theta$', fontsize=14)
    ax[i,0].set_ylabel('$p(\\theta)$', fontsize=14)
```

### Conjugate Prior
A more profund reason for choosing the beta distribution is that it is the *conjugate prior* of the likelihood function - binomial distribution. By using the conjugate prior in Bayesian inference, the posterior is guranteed to be the same function as the prior! There are other pairs of conjugate priors, for example, the Gaussian distribution is the conjugate prior of itself.

### Caculating the posterior
Now that we have defined the likelihood function - binomial distribution and prior - beta distribution, we can use Bayes' rule to caculate the posterior.

$$P(\theta\mid y)={\frac {P(y\mid\theta)\,P(\theta)}{P(y)}}$$

Substituting likelihood $P(y\mid\theta)$ and prior $P(\theta)$, we obtain:
$$P(\theta\mid y)={\frac {{\frac {N!}{N!(N-y)!}}\theta^y(1-\theta)^{N-y}\,{\frac {\Gamma(\alpha + \beta)}{\Gamma(\alpha)\Gamma(\beta)}}\theta^{\alpha-1}(1-\theta)^{\beta-1}}{P(y)}}$$

We obtain $N$ and $y$ from the experiement results. We choose $\alpha$ and $\beta$ to resemble our best knowledge of the prior. The only unknown left is $\theta$; therefore we can drop all other terms that do not dependent on $\theta$.
$$P(\theta\mid y)\propto{\theta^y(1-\theta)^{N-y}\,\theta^{\alpha-1}(1-\theta)^{\beta-1}}$$

Simplifying the expression, we obtain:
$$P(\theta\mid y)\propto\theta^{\alpha-1+y}(1-\theta)^{N-y+\beta-1}$$

Let $\alpha_{posterior}=\alpha_{prior}+y$ and $\beta_{posterior}=\beta_{prior}+N-y$,
$$P(\theta\mid y)\propto\theta^{\alpha_{posterior}-1}(1-\theta)^{\beta_{posterior}-1}$$

We recognize that this posterior distribution is beta distribution. Therefore, we can simply obtain the posterior distribution:
$$P(\theta\mid y) = Beta(\alpha_{prior}+y,\,\beta_{prior}+N-y)$$

### Results
Using the analytical posterior expression we derived, we can plot following posteriors based different priors and experiments. The first plot displays the posteriors with 0 experiments done, hence these curves are just our priors. We have chosen 3 priors for sake of comparison:

* **Blue** is a uniform prior. This is equivalent to saying that all the possible values for the bias are equally probable a priori.
* **Red** is similar to the uniform, but with a little bit more confidence that the bias is either 0 or 1 than the rest of the values.
* **Green** is centered and concentrated around 0.5, so this prior is compatible with information indicating that the coin has more or less about the same chance of landing heads or tails. We could also say this prior is compatible with the belief that most coins are fair.

```python
theta_real = 0.35
trials = [0, 1, 2, 3, 4, 8, 16, 32, 50, 150]
data = [0, 1, 1, 1, 1, 4, 6, 9, 13, 48]

beta_params = [(1, 1), (0.5, 0.5), (20, 20)]
dist = stats.beta
x = np.linspace(0, 1, 100)
plt.subplots(3, 3, sharex=True, sharey=True, figsize=(14, 8))

for idx, N in enumerate(trials):
    if idx == 0:
        plt.subplot(4,3, 2)
    else:
        plt.subplot(4,3, idx+3)
    y = data[idx]
    for (a_prior, b_prior), c in zip(beta_params, ('b', 'r', 'g')):
        p_theta_given_y = dist.pdf(x, a_prior + y, b_prior + N - y)
        plt.plot(x, p_theta_given_y, c)
        plt.fill_between(x, 0, p_theta_given_y, color=c, alpha=0.6)

    plt.axvline(theta_real, ymax=0.3, color='k')
    plt.plot(0, 0, label="{:d} experiments\n{:d} heads".format(N, y), alpha=0)
    plt.xlim(0,1)
    plt.ylim(0,12)
    plt.xlabel(r'$\theta$') 
    plt.legend()
    plt.gca().axes.get_yaxis().set_visible(False)
```

The black vertical line represent the true $\theta$. Of course, we do not know this value in practice. The most probable value is the peak of the posterior distribution (mode).

* The spread of the posterior is proportional to the uncertainty about the value of a parameter.
* Looking at the posteriors for the experiments, we can see that the distriubtions gets tighter as we conduct more tosses. The uncertainty we have in the first example is larger than in the second one because we have more data that supports our inference.
* Given a sufficiently large amount of data, two or more Bayesian models with different priors will tend to converge to the same result. In the limit of infinite number of tosses, no matter which prior we use, we will always get the same posterior.
* How fast posteriors converge to the same distribution depends on the data and the model. In the previous figure we can see that **blue** and **red** posteriors look almost indistinguishable after only 8 experiments, while **green** continues to be separated from the other two even after 150 experiments.
* We can update the posterior sequentially or do it all at once. We can compute the posterior 150 times, each time adding one more observation and using the obtained posterior as the new prior, or we can just compute one posterior for the 150 tosses at once. This feature not only makes perfect sense, also leads to a natural way of updating our estimations when we get new data.
* Evidently, different priors can greatly affect the posterior. Therefore, some people intentionally use non-informative prior. However, if one have reliable prior information, there is no reason to discard that information.

### Bayesian vs Frequentist
Because of the fear that wrong choice of prior can invalidate the inference, maybe it would be better to not have priors at all. This view is frequentist's approach. In fact, many frequentist statistics results can be seen as special cases of a Bayesian model. 

For example, we can see that the mode of the blue posterior equals to the frequentist's expected value, $\hat\theta = \frac{y}{N}$for all experiements. Therfore, by choosing uniform prior in this experiment, Bayesian approach obtains the same result.

Both Bayesian and frequentist approach uses Bayes' rule we talked about earlier. The difference is that Bayesian require a prior, while frequentist only uses likehood function.

Frequentist believe that probability represent long term frequencies of repeatable random experiments. The parameter has a fixed value and it does not make sense to specify a probability distribution for such a parameter. While Bayesians are happy to use probability to describe their incomplete knowledge for a fixed parameter.
